```python
import numpy as np
```

```python
a = np.array([1, 2, 3], dtype='int32')
print(a)
```

```python
b = np.array([[9.0, 8.0, 7.0], [6.0, 5.0, 4.0]])
print(b)
```

```python
# Get Dimensions
print(a.ndim)
print(b.ndim)
```

```python
# Get Shape
print(a.shape)
print(b.shape)
```

```python
# Get Type
print(a.dtype)
print(b.dtype)
```

```python
print(a.itemsize)
print(b.itemsize)
```

```python
# Get total size
print(a.size * a.itemsize)
print(a.nbytes)
```
