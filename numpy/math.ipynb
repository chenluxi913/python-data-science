```python
a = np.ones((2, 3))
print(a)

b = np.full((3, 2), 2)
print(b)

np.matmul(a, b)
```

# Statistics

```python
stats = np.array([[1, 2, 3], [4, 5, 6]])
stats
```

```python
a=np.min(stats)
b=np.min(stats, axis=0)  # min of each column 
c=np.min(stats, axis=1)  # min of each row
print(a)
print(b)
print(c)
```

```python
np.sum(stats)
```
